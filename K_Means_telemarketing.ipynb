```python
import pandas as pd
from sklearn.cluster import KMeans
from matplotlib.pyplot import plot, show

data = pd.read_excel('/content/data telesurvey.xlsx')
```

```python
features = data[['Jumlah Anggota Keluarga', 'Income Code', 'Jumlah_Anak', 'JK_Code', 'USIA', 'RLG_Code', 'U/R Code']]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
k_range = range(1, 4)

wcss_values = []

for k in k_range:

  kmeans = KMeans(n_clusters=k)

  kmeans.fit(scaled_features)

  wcss = kmeans.inertia_
  wcss_values.append(wcss)
```

```python
plot(k_range, wcss_values, marker='o')
plot(k_range, wcss_values, marker='o', linestyle='-')
show()

elbow_point = k_range[wcss_values.index(min(wcss_values))]

print("Nilai K optimal untuk data ini adalah:", elbow_point)
```
